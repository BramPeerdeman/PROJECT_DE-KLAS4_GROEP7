# DWH Aanmaken
Dit notebook maakt de tabellen aan voor het Data Warehouse (DWH) sterschema.

```python
import sqlite3
import os

DWH_PATH = "../data/dwh_deproject.db"

# Verwijder oude database als die bestaat
if os.path.exists(DWH_PATH):
    os.remove(DWH_PATH)
    print(f"Oude database verwijderd: {DWH_PATH}")

conn = sqlite3.connect(DWH_PATH)
cursor = conn.cursor()

sql_statements = [
    """
    CREATE TABLE Dim_Date (
        date_key INTEGER PRIMARY KEY,
        date DATE,
        year INTEGER,
        quarter INTEGER,
        month_name TEXT,
        month_number INTEGER,
        year_month TEXT
    )
    """,
    """
    CREATE TABLE Dim_Staff (
        staff_key INTEGER PRIMARY KEY,
        staff_full_name TEXT,
        position TEXT,
        work_location TEXT
    )
    """,
    """
    CREATE TABLE Dim_Customer (
        customer_key INTEGER PRIMARY KEY,
        company_name TEXT,
        customer_segment TEXT,
        country TEXT,
        region TEXT
    )
    """,
    """
    CREATE TABLE Dim_Product (
        product_key INTEGER PRIMARY KEY,
        product_name TEXT,
        product_type TEXT,
        product_line TEXT,
        introduction_date DATE
    )
    """,
    """
    CREATE TABLE Fact_Sales_Targets (
        date_key INTEGER,
        product_key INTEGER,
        customer_key INTEGER,
        staff_key INTEGER,
        sales_target REAL,
        FOREIGN KEY (date_key) REFERENCES Dim_Date(date_key),
        FOREIGN KEY (product_key) REFERENCES Dim_Product(product_key),
        FOREIGN KEY (customer_key) REFERENCES Dim_Customer(customer_key),
        FOREIGN KEY (staff_key) REFERENCES Dim_Staff(staff_key)
    )
    """,
    """
    CREATE TABLE Fact_Product_Forecast (
        date_key INTEGER,
        product_number INTEGER,
        expected_volume REAL,
        FOREIGN KEY (date_key) REFERENCES Dim_Date(date_key)
    )
    """,
    """
    CREATE TABLE Fact_Sales_Actuals (
        date_key INTEGER,
        product_key INTEGER,
        customer_key INTEGER,
        staff_key INTEGER,
        quantity INTEGER,
        unit_cost REAL,
        unit_price REAL,
        revenue REAL,
        gross_Margin REAL,
        FOREIGN KEY (date_key) REFERENCES Dim_Date(date_key),
        FOREIGN KEY (product_key) REFERENCES Dim_Product(product_key),
        FOREIGN KEY (customer_key) REFERENCES Dim_Customer(customer_key),
        FOREIGN KEY (staff_key) REFERENCES Dim_Staff(staff_key)
    )
    """
]

for stmt in sql_statements:
    cursor.execute(stmt)

conn.commit()
conn.close()
print("DWH tabellen succesvol aangemaakt.")
```
