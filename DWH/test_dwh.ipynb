# DWH Testen
Dit notebook bevat controles om de integriteit van het DWH te verifiëren.

```python
import sqlite3
import pandas as pd

DWH_PATH = "../data/dwh_deproject.db"
conn = sqlite3.connect(DWH_PATH)

def check_table(table_name):
    count = pd.read_sql(f"SELECT COUNT(*) FROM {table_name}", conn).iloc[0,0]
    print(f"{table_name}: {count} records")

tables = [
    "Dim_Date", "Dim_Staff", "Dim_Customer", "Dim_Product",
    "Fact_Sales_Targets", "Fact_Product_Forecast", "Fact_Sales_Actuals"
]

print("Tabel tellingen:")
for t in tables:
    check_table(t)

conn.close()
```

### Steekproef Fact_Sales_Actuals

```python
conn = sqlite3.connect(DWH_PATH)
query = """
SELECT 
    f.*, 
    p.product_name, 
    c.company_name, 
    s.staff_full_name
FROM Fact_Sales_Actuals f
JOIN Dim_Product p ON f.product_key = p.product_key
JOIN Dim_Customer c ON f.customer_key = c.customer_key
JOIN Dim_Staff s ON f.staff_key = s.staff_key
LIMIT 5
"""
sample = pd.read_sql(query, conn)
conn.close()
sample
```
