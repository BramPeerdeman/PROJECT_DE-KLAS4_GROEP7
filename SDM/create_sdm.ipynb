# SDM Aanmaken

```python
import sqlite3
from pathlib import Path

SDM_PATH    = "../data/sdm_deproject.db"
SQL_SCRIPT  = "../data/sdm_create.sql"

# Verwijder de oude database
Path(SDM_PATH).unlink(missing_ok=True)
print(f"Oude database verwijderd: {SDM_PATH}")

# Maak een nieuwe aan met het SQL script
with open(SQL_SCRIPT, "r") as f:
    sql = f.read()

con = sqlite3.connect(SDM_PATH)
con.executescript(sql)
con.commit()

# Bevestiging
tables = con.execute("SELECT name FROM sqlite_master WHERE type='table' ORDER BY name").fetchall()
con.close()

print(f"\n✓ SDM aangemaakt met {len(tables)} tabellen:")
for t in tables:
    print(f"  {t[0]}")
```
